 ***Missing Value***

```python
import pandas as pd
import matplotlib.pyplot as plt


df= pd.read_csv('/content/A.csv')
print(df)



new_df = df.dropna()
print(new_df.to_string())



new_df = df.fillna(99)
print(new_df.to_string())


x = df["weight"].mean()
new_df = df.fillna(x)
print(new_df.to_string())


x = df["weight"].median()
new_df = df.fillna(x)
print(new_df.to_string())


x = df["weight"].mode()[0]
new_df = df.fillna(x)
print(new_df.to_string())


print(df.duplicated())
```
